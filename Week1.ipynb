```python
import numpy as np
import pandas as pd
import cv2 as cv
from google.colab.patches import cv2_imshow # for image display
from skimage import io
from skimage import transform
from PIL import Image
import matplotlib.pylab as plt
```

```python
imgs = ["/content/bunga.jpg",
        "/content/pantai.jpg",
        "/content/kucing.jpg"]

for img in imgs:
  image = io.imread(img)
  image = cv.resize(image,(0,0),fx=0.5, fy=0.5)
  image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
  final_frame = cv.hconcat((image, image_2))
  cv2_imshow(final_frame)
  print('\n')
```

```python
tinggi = image_2.shape[0]
lebar = image_2.shape[1]
print("resolusi image : tinggi x lebar = ", tinggi, "x", lebar)
cv2_imshow(image_2)
```

```python
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

#membuat garis horizontal ditengah image

for y in range (lebar):
  image_3[int((tinggi)/2),y] = [255,255,255]

final_frame = cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)

#
```

```python

image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

tinggi, lebar, _ = image.shape

# Menentukan warna dan ketebalan garis
warna_garis = (255, 255, 255)  # Putih

# Dimulai dari kiri atas ke kanan bawah
cv.line(image_3, (0, 0), (lebar - 1, tinggi - 1), warna_garis)

# Lanjut dari kanan atas ke kiri bawah
cv.line(image_3, (lebar - 1, 0), (0, tinggi - 1), warna_garis)

# Menggabungkan gambar asli dan gambar yang telah dimodifikasi secara horizontal
final_frame = cv.hconcat((image_2, image_3))

# Tampilkan gambar akhir
cv2_imshow(final_frame)
```

```python
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

panjang_garis = 50  #Panjang garis horizontal

# Titik awal dan akhir garis horizontal
start_point = max(0, int((lebar - panjang_garis) / 2))  # Titik awal garis
end_point = min(lebar, start_point + panjang_garis)  # Titik akhir garis
tengah_y = int(tinggi / 2)  # Koordinat y untuk garis horizontal di tengah gambar

# Menambahkan garis horizontal di tengah gambar
for y in range(start_point, end_point):
    image_3[tengah_y, y] = [255, 255, 255]

final_frame = cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)
```

```python

image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = image_2.copy()
kotak_lebar = min(50, lebar)
kotak_tinggi = min(50, tinggi)

start_x = max(0, int((lebar - kotak_lebar) / 2))
start_y = max(0, int((tinggi - kotak_tinggi) / 2))


for y in range(start_y, start_y + kotak_tinggi):
    for x in range(start_x, start_x + kotak_lebar):
        image_3[y, x] = [255, 255, 255]  # Warna putih

final_frame = cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)


```
