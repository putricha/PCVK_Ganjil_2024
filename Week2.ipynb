```python
image = cv.imread('/content/drive/MyDrive/PCVK/kucing.jpg')
blue_channel, green_channel, red_channel = cv.split(image)

# Membuat gambar yang hanya menampilkan saluran biru
only_blue_image = cv.merge([blue_channel, blue_channel * 0, blue_channel * 0])
only_blue_image = cv.cvtColor(only_blue_image, cv.COLOR_BGR2RGB)

# Menampilkan gambar hanya dengan saluran biru

plt.title('Blue Channel')
plt.imshow(only_blue_image)
```

```python
cv.rectangle(only_blue_image, pt1=(0,120), pt2=(130,160), color=(0,200,0), thickness=5)
plt.imshow(only_blue_image)
```
